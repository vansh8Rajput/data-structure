Q8 What is a hash table, and how does it relate to dictionaries in Python?
Ans: Hash table is used to store the value in key-value pair.
we can access the value useing the key inside hash table.
It uses a hash function to convert each key into a unique index, where the corresponding value is stored in memory.
In Python, the dictionary (dict) is implemented using a hash table. This design makes accessing, adding, and removing values by key extremely fast. — usually in O(1) time.
for example:
# dictionary using a hash table
student = {'name': 'Vansh', 'age': 21, 'grade': 'A'}
# Accessing a value
print(student['name'])  # Output: Vansh

Q16. What is a nested dictionary, and give an example of its use case?
Ans:A nested dictionary in Python is a dictionary inside another dictionary.
This structure allows you to store multiple levels of related data in a clear and organized way.
For example:
students = {
    '101': {'name': 'Vansh', 'age': 21, 'grade': 'A'},
    '102': {'name': 'Aryan', 'age': 22, 'grade': 'B'}
}
print(students['101']['name'])  #output: Vansh
print(students['102']['grade']) #Output: B
